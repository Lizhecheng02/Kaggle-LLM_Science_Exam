```python
import pandas as pd
import regex as re
import os
import random
```

```python
df = pd.DataFrame()

for file_name in os.listdir("./Raw Dataset/70k new dataset"):
    if file_name.endswith("json"):
        print("file path:", os.path.join(
            "./Raw Dataset/70k new dataset", file_name))
        tmp_df = pd.read_json(os.path.join(
            "./Raw Dataset/70k new dataset", file_name))
        df = pd.concat([df, tmp_df])

print(df.shape)

df = df.reset_index(drop=True)
df_shuffle = df.copy()


def shuffle_answer(idx):
    try:
        tmp = ['A', 'B', 'C', 'D', 'E']
        random.shuffle(tmp)
        hm = {k: v for k, v in zip(['A', 'B', 'C', 'D', 'E'], tmp)}
        hm_b = {v: k for k, v in zip(['A', 'B', 'C', 'D', 'E'], tmp)}
        df_shuffle.loc[idx, ['A', 'B', 'C', 'D', 'E']] = [
            df.loc[idx, hm[k]] for k in ['A', 'B', 'C', 'D', 'E']]
        df_shuffle.loc[idx, 'answer'] = hm_b[df.loc[idx, 'answer']]
    except:
        print("Error!")


for i in range(len(df_shuffle)):
    shuffle_answer(i)

df = df_shuffle.copy()
print(df["answer"].value_counts())
print(df.shape)

correct_answer_format = ["A", "B", "C", "D", "E"]
df = df[df["answer"].isin(correct_answer_format)]
print(df.shape)

df_with_nan = df[df.isna().any(axis=1)]
print(df_with_nan.shape)

df = df.drop_duplicates(subset=["prompt"])
print(df.shape)


def clean_string(s):
    text = re.sub(r'#+', '', s)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    text = re.sub(r'\(\)', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


df["context"] = df["context"].apply(clean_string)

df.to_json("./Clean Dataset/58k no #.json", orient="records")
```
