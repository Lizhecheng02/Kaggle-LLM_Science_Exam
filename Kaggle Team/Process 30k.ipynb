```python
import pandas as pd
```

```python
import os

df_17k = pd.DataFrame()

for file_name in os.listdir("./raw dataset/"):
    if file_name.startswith("17k"):
        print("file path:", os.path.join("./raw dataset/", file_name))
        tmp_df = pd.read_json(os.path.join("./raw dataset/", file_name))
        df_17k = pd.concat([df_17k, tmp_df])

print(df_17k.shape)
```

```python
df_17k.head()
```

```python
df_17k_with_nan = df_17k[df_17k.isna().any(axis=1)]
print(df_17k_with_nan.shape)
```

```python
df_17k.rename(columns={"question": "prompt", "top_best_answer": "answer"}, inplace=True)
df_17k.head()
```

```python
df_13k = pd.DataFrame()

for file_name in os.listdir("./raw dataset/"):
    if file_name.startswith("train_13k"):
        print("file path:", os.path.join("./raw dataset/", file_name))
        tmp_df = pd.read_json(os.path.join("./raw dataset/", file_name))
        df_13k = pd.concat([df_13k, tmp_df])

print(df_13k.shape)
```

```python
df_13k.head()
```

```python
df_13k.drop(columns=["answer_all", "prompt_answer_stem"], axis=1, inplace=True)
df_13k.head()
```

```python
column_order = ["question", "context", "A", "B", "C", "D", "E", "answer"]
df_13k = df_13k[column_order]
df_13k.head()
```

```python
df_13k.rename(columns={"question": "prompt"}, inplace=True)
df_13k.head()
```

```python
df_30k = pd.concat([df_17k, df_13k])
df_30k.head()
```

```python
df_30k_with_nan = df_30k[df_30k.isna().any(axis=1)]
print(df_30k_with_nan.shape)
```

```python
import regex as re

def clean_string(s):
    text = re.sub(r'#+', '', s)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    text = re.sub(r'\(\)', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def use_sep(s):
    text = re.sub(r'#####', '####', s)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    text = re.sub(r'\(\)', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
df_30k["context"] = df_30k["context"].apply(use_sep)
df_30k.to_json("./final dataset/30k full with #.json", orient="records")
```
