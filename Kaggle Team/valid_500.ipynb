```python
import pandas as pd
```

```python
valid_300 = pd.read_json("./Raw Dataset/gpt4-correct-300.json")
# valid_300.insert(0, "id", range(len(valid_300)))
valid_300.head()
```

```python
valid_200 = pd.read_json("./Clean Dataset/200 val no #.json")
valid_200.head()
```

```python
valid_500 = pd.concat([valid_300, valid_200])
valid_500.head()
```

```python
import regex as re

def clean_string(s):
    text = re.sub(r'#+', '', s)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    text = re.sub(r'\(\)', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

valid_500["context"] = valid_500["context"].apply(clean_string)

valid_500.to_json("./Clean Dataset/500 val gpt4 no #.json", orient="records")
```
