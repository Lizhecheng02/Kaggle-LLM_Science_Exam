```python
"""
Download latest Wikipedia Dataset from huggingface
"""

from datasets import load_dataset
import pandas as pd

wikipedia_en = load_dataset("graelo/wikipedia", "20230901.en")
print(wikipedia_en)

batch_size = 200_000 
num_batches = len(wikipedia_en["train"]) // batch_size + 1

for i in range(num_batches):
    max_index = min((i + 1) * batch_size, len(wikipedia_en["train"]))
    batch = wikipedia_en["train"].select(range(i * batch_size, max_index))
    batch_df = pd.DataFrame(batch)
    batch_df.to_parquet(f"wikipedia_dataset_batch_{i + 1}.parquet")
    print(f"Successfully save index from {i * batch_size} to {max_index - 1}")
```
